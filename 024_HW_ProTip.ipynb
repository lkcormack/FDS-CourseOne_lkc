# Pro Tip: Generating custom sinusoids

To generate a sinewave, we need a function to compute the sine (or consine), and a series of numbers over which to compute the sine. `Numpy` is our friend.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 100)
y = np.sin(t)
```

Now let's look at our creation:

```python
plt.plot(t, y)
```

Notice that the `sin()` function takes an argument in *radians*. Thus, when we gave it values of `t` that ran from 0 to 2\**pi*, we got exactly one cycle of the sine wave. But what if we want `t` to be in another unit, like *time* in *seconds*?  

To do this, we just need to add a `2\**pi*` to our argument to the `sin()` function. This way any one-unit change in t will result in a one-cycle progression of the sine wave:

```python
y = np.sin(2*np.pi*t)
plt.plot(t, y)
```

Do you see what happened? Now a change of 1 in `t` corresponds to a complete cycle. Thus, as `t` goes from 1 to 6 (well, 6.28...), we get about 6 cycles of a sine wave.

Now for the fun part. Since we have our units sorted, we can add a couple additional parameters to adjust the *frequency* and *phase* of our oscillation. 

Let's start with frequency, which is how many cycles we want per unit of `t`. We'll donote this with a `w`, which is close the the little omega generally used for frequency in maths.

```python
w = 0.5       # frequency (cyclese per unit time)
y = np.sin(2*np.pi*w*t)
plt.plot(t, y)
```

That gives us 1/2 a cycle per unit `t`. We can set *`w`* to whatever we want to control the frequency of oscillation.

Now let's play with the phase. To make the phase shift clear, we'll make and plot 2 sinusoids: one with a phase shift and one without. We'll donate phase with a `p` (for *phi*).

```python
w = 0.5                         # frequency in cycles per your unit of time
p = np.pi/2                     # phase shift in radians
y = np.sin(2*np.pi*w*t)
y2 = np.sin(2*np.pi*w*t + p)
plt.plot(t, y)
plt.plot(t, y2)
```

Notice the phase shift here is expressed in radians. It's usually more convenient to express the shift in whatever unit we're working in, such as hours or days or whatever. To do this, we just need to apply shift directly to the `t` variable:

```python
w = 0.5                         # frequency in cycles per your unit of time
p = 0.2                        # phase shift in your unit of time
y = np.sin(2*np.pi*w*t)
y2 = np.sin(2*np.pi*w*(t + p))
plt.plot(t, y)
plt.plot(t, y2)
```

So there we have it! Now we can make sinusoidal oscillations in whatever unit we want (e.g. hours, days, etc.), specify the frequency of those oscillations (e.g. cycles per day), and the phase shift (e.g. fraction of a day).
